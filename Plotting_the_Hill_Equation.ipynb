# The Hill Equation

We have discussed the use of the Hill equation a general
functional form that can be used to describe *cooperativity* in
the rate of biochemical reactions:
$$
\nu = \frac{V_{\rm max}[S]^h}{K_{1/2}^h + [S]^h}
$$

For enzyme-catalyzed reactions, the Hill equation becomes identical
to the Michaelis-Menten equation when $h=1$, and when derived from first-principles $V_{\rm max} = k_{\rm cat}E_{\rm tot}$ (the product of the rate of the catalytic step and the total concentration of the enzyme).

## Varying the Hill coefficient.

Create a single graph that plots the the rate vs substrate concentration over a range of $[S]=0$ to $100$ mM for $h=1$, $2$ and $10$, with $V_{\rm max}$ and $K_{1/2}$ held constant at $V_{\rm max} = 5.0$ mM/s and $K_{1/2}=20.0$ mM.


Visually explore how the parameters of the Hill equation impact its shape by creating plots of the equation with varying parameters.
\begin{itemize}
    \item Visually explore how the parameters of the Hill equation impact its shape by creating plots of the equation with varying parameters.  Create (computationally) three individual graphs that plot the rate vs substrate concentration over a range of $[S]=0$ to $100$~mM with a single varying parameter:




```python
# Enter code to vary h

import numpy as np
from matplotlib import pyplot as plt

s = np.linspace(0, 100, 100)
v_max = 5
k_half = 20

h = 1
v = v_max * s**h / (k_half**h + s**h)
plt.plot(s, v)

h = 2
v = v_max * s**h / (k_half**h + s**h)
plt.plot(s, v)

h = 10
v = v_max * s**h / (k_half**h + s**h)
plt.plot(s, v)

plt.title("Hill Plot for h = 10")
plt.xlabel("Substrate Concentration [S]")
plt.ylabel("Reaction Velocity (v)")
plt.grid(True)
plt.show()
```

## Varying $V_{max}$

Create a single graph that plots the the rate vs substrate concentration over a range of  [S]=0  to  100  mM for $V_{\rm max}=2.0$, $5.0$ and $10.0$ mM/s, with $h$ and $K_{1/2}$ held constant at $h=2$ and $K_{1/2}=20.0$ mM.

```python
# Enter code to vary V_max

import numpy as np
from matplotlib import pyplot as plt

s = np.linspace(0, 100, 100)
h = 2
k_half = 20

v_max = 2
v = v_max * s**h / (k_half**h + s**h)
plt.plot(s, v)

v_max = 5
v = v_max * s**h / (k_half**h + s**h)
plt.plot(s, v)

v_max = 10
v = v_max * s**h / (k_half**h + s**h)
plt.plot(s, v)
plt.title("Hill Plot for v_max = 10")
plt.xlabel("Substrate Concentration [S]")
plt.ylabel("Reaction Velocity (v)")
plt.grid(True)
plt.show()
```
